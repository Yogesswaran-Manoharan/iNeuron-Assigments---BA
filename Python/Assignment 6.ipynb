```python
# Q1 Python program to add two Matrices

# Get Input from User 

# Get the matrix shape from user for both Matrix 1 and Matrix 2
# Since both matrix should be same shape, We can use one input for creating shape of both matrix

row  = int(input("Enter the row size :"))
column = int(input("Enter the column size :"))

# Now create the nested list
print("Enter the values for Matrix1")
# We are going to use list comprehension method
# Matrix 1
Matrix1 = [[int(input()) for i in range(column)] for j in range(row)] 

# For printing the Matrix1 Values in matrix form
print("Matrix1 :")
for i in range(row):
    for j in range(column):
        print(format(Matrix1[i][j],"<3"),end=" ")
    print()

#Matrix 2
print("Enter the values for Matrix2")
Matrix2 = [[int(input()) for i in range(column)] for j in range(row)] 

# For printing the Matrix1 Values in matrix form
print("Matrix2 :")
for i in range(row):
    for j in range(column):
        print(format(Matrix2[i][j],"<3"),end=" ")
    print()

# Calculation

# Create nested list for adding two matrices "Since we need empty nested list keep initial value 0"
result = [[0 for i in range(column)] for j in range(row)] 
# create loop
for i in range(row):
    for j in range(column):
        result[i][j] = Matrix1[i][j] + Matrix2[i][j]

# Output

# Print the final added values of both matrices
print("Added value of two matrices :")
for i in range(row):
    for j in range(column):
        print(format(result[i][j],"<3"),end=" ")
    print()
```

```python
# Q2 Python program to multiply two matrices

# Get Input from User 

# Get the matrix shape from user for both Matrix 1 and Matrix 2
# Since Matrix 1 column size and matrix 2 row size should be equal, we can get only three inputs
# 1. Row size for M1 2. Column size for M 2 and 3. Common size Col M1/ Row M2

p  = int(input("Enter the row size for M1:"))
q = int(input("Enter the column size for M2:"))
n = int(input("Enter the column size for col_M1/row_M2:"))

# Now create the nested list
print("Enter the values for Matrix1")
# We are going to use list comprehension method
# Matrix 1
Matrix1 = [[int(input()) for i in range(n)] for j in range(p)] 

# For printing the Matrix1 Values in matrix form
print("Matrix1 :")
for i in range(p):
    for j in range(n):
        print(format(Matrix1[i][j],"<5"),end=" ")
    print()

#Matrix 2
print("Enter the values for Matrix2")
Matrix2 = [[int(input()) for i in range(q)] for j in range(n)] 

# For printing the Matrix1 Values in matrix form
print("Matrix2 :")
for i in range(n):
    for j in range(q):
        print(format(Matrix2[i][j],"<5"),end=" ")
    print()

# Calculation

# Create nested list for adding two matrices "Since we need empty nested list keep initial value 0"
result = [[0 for i in range(q)] for j in range(p)] 
# create loop
for i in range(p):
    for j in range(q):
        for k in range(n):
            result[i][j] = result[i][j] + Matrix1[i][k] * Matrix2[k][j]

# Output

# Print the final added values of both matrices
print("Multiplied value of two matrices :")
for i in range(p):
    for j in range(q):
        print(format(result[i][j],"<5"),end=" ")
    print()
```

```python
# Q5 Transpose a matrix in Single line in Python

# Input
row  = int(input("Enter the row size :"))
column = int(input("Enter the column size :"))

print("Enter the values for Matrix1")
# Matrix 1
Matrix1 = [[int(input()) for i in range(column)] for j in range(row)] 

print("Matrix :")
for i in range(row):
    for j in range(column):
        print(format(Matrix1[i][j],"<5"),end=" ")
    print()
    
# Transpose
result = [[0 for i in range(row)] for j in range(column)]

# Output

for i in range(column):
    for j in range(row):
        result[i][j] = Matrix1[j][i]

print("Transposed Matrix :")
for i in range(column):
    for j in range(row):
        print(format(result[i][j],"<5"),end=" ")
    print()
```

```python
# Q8 Python â€“ Vertical Concatenation in Matrix

# Input
x = [['We ',' to '],[' are ',' movie '],[' going ',' today.']]

# Using loop

lis = []
n = 0
while n != len(x):
    temp = ''
    for i in x:
        # check for valid index/column
        try: temp = temp + i[n]
        except IndexError: pass
    lis.append(temp)
    n = n+1

# Outputx`
lis = [ l for l in lis if l]

print("List after column concatenation : ",lis)
```

```python
# Q10 Python program to check whether the string is Symmetrical or Palindrome

def palin(string): 
  # declare and initialize with the starting and ending indexes
  st = 0
  end = len(string)-1
  f = 0
  # loop comparing letters moving from start to end and from end to start
  while(st<end): 
    if (string[st]== string[end]): 
      
      st += 1
      end -= 1
      
    else: 
      f = 1
      break; 
  # if loop with f as condition    
  if f == 0: 
    print("The entered string is palindrome") 
  else: 
    print("The entered string is not palindrome") 
    
# symm function to check string symmetrical or not
def symm(string): 
  
  l = len(string) 
  flag = 0
  
  # to check length of string even or odd 
  # to calculate middle value accordingly
  if l%2 == 0: 
    mid = l//2 # for even length
  else: 
    mid = l//2 + 1 # for odd length
    
  s1 = 0  # starting for first portion of string
  s2 = mid # starting for rest portion of string after middle value
  
  while(s1 < mid and s2 < l): 
    
    if (string[s1] == string[s2]): # comparing from start of both portions 
                                  # of given string
      s1 = s1 + 1
      s2 = s2 + 1
    else: 
      flag = 1
      break
  
  if flag == 0: 
    print("The entered string is symmetrical") 
  else: 
    print("The entered string is not symmetrical") 
    
# Main code 
string = input("Enter the string: ")
palin(string) 
symm(string)
```
